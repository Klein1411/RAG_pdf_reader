```python
# 1) Nạp module và thêm project folder (nếu cần)
import sys
sys.path.append(r"D:\Project_self")

from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
from langchain.chains import RetrievalQA
from langchain.llms import Ollama
```

```python
llm = Ollama(model="gpt-oss:20b")
```

```python
# 2) Load lại vector store offline
hf_emb = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vectordb = Chroma(
    persist_directory="./chromadb_offline",
    embedding_function=hf_emb
)
retriever = vectordb.as_retriever(search_kwargs={"k": 4})
```

```python
# 4) Tạo chain RetrievalQA
qa = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriever,
    return_source_documents=True
)
```

```python
# 5) Chạy truy vấn
question = "What is the main objective of this research project?"
result = qa(question)

print("=== ANSWER ===")
print(result["result"])

print("\n=== SOURCES ===")
for doc in result["source_documents"]:
    page = doc.metadata.get("page", "unknown")
    snippet = doc.page_content.replace("\n", " ").strip()[:200]
    print(f"- (page {page}) {snippet}…")
```
